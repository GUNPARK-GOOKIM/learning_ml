```python
s = pd.read_csv('C:\\Users\\Lenovo\\house_prices.csv')
```

```python
s.drop(columns=['Dimensions','Plot Area'], inplace = True)
```

```python
s.drop(columns = ['facing'], inplace = True)
```

```python
s.head()
```

```python
s.fillna({"Furnishing":s["Furnishing"].mode()[0]}, inplace = True)
s.fillna({"Balcony": 0}, inplace=True)
s["Bathroom"] = pd.to_numeric(s["Bathroom"], errors="coerce")
s["Bathroom"] = s["Bathroom"].fillna(s["Bathroom"].median())


s.fillna({"Status":"Unknown"}, inplace = True)
s.fillna({"Ownership":"Unknown"}, inplace = True)
s.fillna({"Transaction":s["Transaction"].mode()[0]}, inplace = True)
s.fillna({"Price (in rupees)":s["Price (in rupees)"].mean()}, inplace = True)
s.fillna({"Floor":"Unknown"}, inplace = True)
```
